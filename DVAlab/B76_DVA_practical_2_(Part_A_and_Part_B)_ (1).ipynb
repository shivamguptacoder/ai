PART (B): Data Preprocessing
Business challenge/requirement
Perform data wrangling for studentâs data. The data is maintained for subjects: Math, Physics and Data
Structure. This data is stored in different files. Task is to perform analysis of scores and find if there is
any correlation of score with age, ethnicity etc.

DO as directed:


```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read the three csv files which contains the score of same students in term1 for each Subject


df_math = pd.read_csv("B_76_ShivamGupta - MathScoreTerm1.csv")
df_dss = pd.read_csv("B_76_ShivamGupta - DSScoreTerm1.csv")
df_phy = pd.read_csv("B_76_ShivamGupta - PhysicsScoreTerm1.csv")


print(df_math.head())
print('\n')
print(df_dss.head())
print('\n')
print(df_phy.head())

# INTERPRETATION
# The following shows the first 5 tuples of each dataset  of mathscore Physicsscore and dsscore
```

```python
# 2. Remove the name and ethnicity column (to ensure confidentiality)
# Delete Ethinicity
del df_math["Ethinicity"]
del df_dss["Ethinicity"]
del df_phy["Ethinicity"]

# Delete Name
del df_math["Name"]
del df_dss["Name"]
del df_phy["Name"]

print(df_math.head())
print('\n')
print(df_dss.head())
print('\n')
print(df_phy.head())

# INTERPRETATION
# The following shows the first 5 tuples of each dataset  of mathscore Physicsscore and dsscore
# after removing name and ethinicity column
```

```python
# 3. Fill missing score data with zero
df_math.fillna(0)
df_dss.fillna(0)
df_phy.fillna(0)

print(df_math.head())
print('\n')
print(df_dss.head())
print('\n')
print(df_phy.head())

# INTERPRETATION
# This is the Preprocessing of data part where missing data is replaced with zero
```

```python
# 4. Merge the three files
df_merged = df_math.merge(df_dss, on = "ID", suffixes = ('math', 'ds')).merge(df_phy, on = "ID", suffixes = ("ds", "physics"))

merged_df = df_merged.filter(["ID", "Score_math", "Score_ds", "Score", "Age_math", "Sex", "Ethinicity"]).rename(columns={"Score": "Score_physics", "Age_math": "Age"})



# INTERPRETATION
# here we merge 3 files 
```

```python

# 5. Change Sex(M/F) Column to 1/2 for further analysis
merged_df["Sex"] = [1 if sex == "M" else 2 for sex in merged_df["Sex"]]

print(merged_df)


# INTERPRETATION
# Here we channge SEX from male to female 
```

```python
#6. Store the data in new file âScoreFinal.csv
merged_df.to_csv("ScoreFinal.csv")


# INTERPRETATION
# merged DF is saved to Scorefinal.csv file
```

```python
#7. Convert ethnicity to numerical value

mapping = {"White American" : 1, "European American" : 2, "Hispanic" : 3, "African American" : 4}

merged_df.replace({'Ethinicity' :  mapping})


# INTERPRETATION
# The ethinicity is converted to numerical value
```

```python
#8. Fill the missing score for a student to the average of the class

mean_score = merged_df["Score_physics"].mean()
merged_df['Score_physics'].fillna(merged_df['Score_physics'].mean(), inplace=True)

print(merged_df)

# INTERPRETATION
# We have succesfully filled the missing scores with mean of class
```
