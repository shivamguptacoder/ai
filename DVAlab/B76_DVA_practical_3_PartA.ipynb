**Read the file and drop all the nan values**

```python
df=pd.read_excel("B_34_Nidhi Misra - online_retail_II.xlsx")
df
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
df_plus=df[df['Quantity']>=1]
df_plus.info()
```



---


### **a.** **Create basket data using the transactions from Germany, encode it and consider only the transactions with more than 1 item only**.

```python
basket_germany=(df_plus[df_plus['Country']=="Germany"].groupby(['Invoice','Description'])['Quantity']
             .sum().unstack().reset_index().fillna(0).set_index('Invoice'))
basket_germany
```

```python
def encode_units(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
basket_encode_germany=basket_germany.applymap(encode_units)
basket_encode_germany
```

```python
basket_filter_germany=basket_encode_germany[(basket_encode_germany>0).sum(axis=1)>=2]
basket_filter_germany
```



---


### **b. Apply the apriori algorithm and find associations between the frequently bought items. Conclude which items are bought together.**

```python
frequent_itemsets_germany=apriori(basket_filter_germany,min_support=0.03,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
frequent_itemsets_germany['length']=frequent_itemsets_germany['itemsets'].apply(lambda x:len(x))
frequent_itemsets_germany
```

```python
association_rules(frequent_itemsets_germany,metric='confidence',min_threshold=0.02).sort_values('lift',ascending=False).reset_index(drop=True)
```

**Analysis:**
  From the above data, we can conclude that if a customer buys **SPACEBOY CHILDRENS BOWL** then there is high probability that he/she will also buy **SPACEBOY CHILDRENS CUP**, since the confidence and lift values are **0.894737** and **19.371053** repectively
