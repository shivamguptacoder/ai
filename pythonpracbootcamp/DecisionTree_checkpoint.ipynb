```python
import math
import csv

def entropyofclass(A):
    p=0
    n=0
    total=0
    for i in range(len(A)):
        if(A[i]=='yes'):
            p+=1  
        else:
            n+=1
        total+=1
    x=0
    try:
        x=x-((p/total)*math.log((p/total),2))
    except:
        x=x-0
    try:
        x=x-((n/total)*math.log((n/total),2))
    except:
        x=x-0
    return x

def entropyofattribute(A,B,string):
    p=0
    n=0
    total=0
    # print(A)
    for i in range(len(A)):
        if(A[i]==string and B[i]=='yes'):
            p+=1 
            total+=1
        elif(A[i]==string and B[i]=='no'):
            n+=1
            total+=1
    # print(p,n,total)
    x=0
    try:
        x=x-((p/total)*math.log((p/total),2))
    except:
        x=x-0
    try:
        x=x-((n/total)*math.log((n/total),2))
    except:
        x=x-0
    return x,total
        





c=0
ll=[]
list1=[]
with open('playTennis.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        if(c==0):
            ll=row[0].split(",")
            list1=[[] for i in range(len(ll))]
            for i in range(len(ll)):
                list1[i].append(ll[i])
            c+=1
        else:
            ll=row[0].split(",")
            for i in range(len(ll)):
                list1[i].append(ll[i])
                
                
dict1={}
for i in range(len(list1)):
    dict1[list1[i][0]]=list1[i][1:]

# print(dict1)
entropy1=entropyofclass(dict1['PlayTennis'])

dict2={}
for i in dict1:
    if(i!="Day" and i!="PlayTennis"):
        set1=set()
        total=0
        for j in dict1[i]:
            set1.add(j)
            total+=1
        ig=entropy1
        for k in set1:
            # print(k,dict1[i])
            entropy,t=entropyofattribute(dict1[i],dict1['PlayTennis'],k)
            # list2.append()
            ig-=(t/total)*entropy
        dict2[i]=ig

ss=""
c=0
for i in dict2:
    if(dict2[i]>c):
        c=dict2[i]
        ss=i
print("Root is "+ss)
print()

x=set(dict1['Outlook'])

for k in x:
    dict11={}
    for i in dict1:

        dict11[i]=[]
        for j in range(len(dict1[i])):
            if(dict1['Outlook'][j]==k):
                dict11[i].append(dict1[i][j])
        
    dict2={}   
    for i in dict11:
        if(i!="Day" and i!="playtennis" and i!="Outlook"):
            set1=set()
            total=0
            for j in dict11[i]:
                set1.add(j)
                total+=1
            ig=0.971
            for kk in set1:
               
                entropy,t=entropyofattribute(dict11[i],dict11['PlayTennis'],kk)
                # list2.append()
                ig-=(t/total)*entropy
            dict2[i]=ig

    print(k,dict2,)


```
